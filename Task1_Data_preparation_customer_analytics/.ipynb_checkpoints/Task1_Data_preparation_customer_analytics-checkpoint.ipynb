```python
transactionData = pd.read_excel("QVI_transaction_data.xlsx")
customerData = pd.read_csv("QVI_purchase_behaviour.csv")
```

```python
df1=transactionData
df2=customerData
# Check transactionData Outliers, nulls and duplicated.
print('''transactionData contain %s rows \n\nNull data counts: %s\n\noutlier data include: \n 
LYLTY_CARD_NBR outliers count: %s \nTXN_ID outliers count: %s
PROD_QTY outliers count: %s \nTOT_SALES outliers count: %s
Duplicated values: %s \nDATE column data type is: %s ,need to convert it to Datetime.
\n**Will Drop outlier, duplicated and correct Date type.**\n\n''' %(len(df1),df1.isnull().sum().sum(),
                            df1.index[(df1['LYLTY_CARD_NBR']>500000)].size,
                            df1.index[(df1['TXN_ID']>500000)].size,
                            df1.index[(df1['PROD_QTY']>5)].size,
                            df1.index[(df1['TOT_SALES']>100)].size,
                            df1.index[df1.duplicated()].size,
                            df1['DATE'].dtype))

#Drop transactionData Outliers, nulls and duplicated.
df1 = df1.drop(df1.index[df1.duplicated()])
df1 = df1.drop(df1.index[(df1['LYLTY_CARD_NBR']>500000) | (df1['TXN_ID']>500000) | (df1['PROD_QTY']>5) | (df1['TOT_SALES']>100)])
df1['DATE'] = pd.to_datetime(df1['DATE'],unit='D', origin='1899-12-30')

# Check customerData Outliers, nulls and duplicated.
print('''=========\n\ncustomerData contain %s rows \n\nNull data counts: %s\n\noutlier data include: \n
LYLTY_CARD_NBR outliers count: %s \nNo Duplicated values Found.\n''' %(len(df1),df2.isnull().sum().sum(),
                                                                        df2.index[(df2['LYLTY_CARD_NBR']>500000)].size))
#Drop customerData Outliers, nulls and duplicated.
df2 = df2.drop(df2.index[df2['LYLTY_CARD_NBR']>500000])

#Show data infos
print('\n\ntransactionData info:\n\n')
df1.info()
print('\n\ncustomerData info:\n\n')
df2.info()
```

```python
print('We will remove salsa product and focus on chips products\n\nsalsa product are: \n')
salsaproduct = []
for x in df1.groupby('PROD_NAME').count().index:
    if re.search("salsa", x.lower()):
        salsaproduct.append(x)
        print(x)
for x in salsaproduct:
    df1.drop(df1.index[df1['PROD_NAME']=='Doritos Salsa       Medium 300g'],inplace=True)
```

```python
# UniqueWordslist = pd.DataFrame([y for x in df1['PROD_NAME'] for y in x.split(' ')])
# cleanUniqueWordslist = re.sub(r"[123456789@#$!%^&*()-=_+?~`<>{}/\؟':;,.؛،’]+", '', ' '.join([x for x in UniqueWordslist[0].tolist()]), flags=re.UNICODE)
# cleanUniqueWordslist = re.sub(' g ', ' ', cleanUniqueWordslist)
# cleanUniqueWordslist = cleanUniqueWordslist.split(' ')
# cleanUniqueWordslist = pd.DataFrame(cleanUniqueWordslist,columns=['Words'])
# np.array(set(cleanUniqueWordslist['Words']))
# cleanUniqueWordslist['Words'].value_counts()
print('Adding new column for Clean_PROD_NAME.\n\nSample:')
CleanProductList = []
for x in df1['PROD_NAME']:
    clean = re.sub(r"[123456789@#$!%^&*()-=_+?~`<>{}/\؟':;,.؛،’]+", ' ',x, flags=re.UNICODE)
    clean = re.sub(' g', '', clean)
    clean = re.sub('  +', ' ', clean)
    clean = re.sub(' +$', '', clean)
    CleanProductList.append(clean)
df1['Clean_PROD_NAME']=CleanProductList
CleanProductListtable = pd.DataFrame(CleanProductList,columns=['Clean_PROD_NAME']).reset_index()
random.choices(list(set(df1['Clean_PROD_NAME'])),k=10)
```

```python
testdate = df1.groupby('DATE').count().reset_index()
testdate['day'] = testdate['DATE'].dt.day 
testdate['month'] = testdate['DATE'].dt.month

testdate.plot(x='DATE',y='TXN_ID',figsize=(20,10))
plt.title('graph show strange movment on December lets try to check transactions count per day',fontsize=20)
plt.show()
testdate.groupby('day').count().reset_index().plot.bar(x='day',y='TXN_ID',figsize=(15,5))
plt.title('transactions count per day show decrease in 25 and after 29 of December',fontsize=20)
plt.show()
mis = pd.date_range(start = '2018-12-01', end = '2018-12-31' ).difference(testdate['DATE'][(testdate['DATE']<='2018-12-31') & (testdate['DATE']>='2018-12-01')])
print('\n\nAfter checking if there are any missing date in December...Opps %s is missing!!\nIts zero sales on Christmas day so nothig abnormal\nPlus to the decrease of slaes on the end of month 29,30,31' %(mis.date))
```

```python
print('Adding new column for Pack_size.\n\nPack sizes available:')
Pack_size = []
for x in df1['PROD_NAME']:
    if re.search(" *g", x.lower()):
        Pack_size.append(re.search(r"\d+g", x.lower()).group(0))
df1['Pack_size'] = Pack_size
np.array(set(df1['Pack_size']))
```

```python
combineSameBranddic = {'WW':'Woolworths',
                       'NCC':'Natural',
                       'Dorito':'Doritos',
                       'RRD':'Red'}
print('Adding new column for Brand_Name\n\nCombine Same Brand to Unify Brand Name:\n%s\n\nBrand Names available:' %(combineSameBranddic))
Brand_Name = []
for x in df1['PROD_NAME']:
    Brand_Name.append(x.split(' ')[0])
df1['Brand_Name'] = Brand_Name
df1['Brand_Name'].replace(combineSameBranddic, inplace=True)
np.array(set(df1['Brand_Name']))
```
